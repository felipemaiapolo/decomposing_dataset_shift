```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.feature_selection import SelectFromModel

import xgboost as xgb

pd.set_option('display.max_columns', None)
```

```python
random_state=42
```

```python
src = pd.read_csv("data/dataset_t0_complete.csv") 
src.head(5)
```

```python
src.shape
```

```python
X = src.drop(columns=['SConcEver30dP3_100','ndoc','Unnamed: 0','autoriza_pos_b2c','autoriza_pos','autoriza_pos_b2b','zmodultcontratoativo'])
y = src.loc[:,'SConcEver30dP3_100']
ids = src.loc[:,'ndoc']
```

```python
X = pd.DataFrame(X)
X = X.astype(float)

y = pd.DataFrame(y)

X.shape, y.shape
```

# Feature Importance

```python
y = y.squeeze()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
```

```python
%%time

np.random.seed(random_state)

model = xgb.XGBClassifier(n_estimators=3000, 
                          n_jobs = -1,
                          seed=random_state,
                          random_state=random_state)

model.fit(X_train, y_train,
          verbose=True,
          early_stopping_rounds=100,
          eval_metric='auc',
          eval_set=[(X_test, y_test)])
```

```python
#make predictions for test data
y_pred = model.predict_proba(X_test)[:,1]

#evaluate predictions
auc = roc_auc_score(y_test, y_pred)
print("AUC: %.5f%%" % (auc * 100.0))
```

```python
importance_dict = model.get_booster().get_score(importance_type='gain')

feat=np.array(list(importance_dict.keys()))
imp=np.array(list(importance_dict.values()))

imp.shape[0]
```

```python
ind=np.argsort(imp)[::-1]
d=75 #select only the d most important

feat=feat[ind][:d].tolist()
imp=imp[ind][:d]
```

```python
with open("data/importance_names.txt", "wb") as fp:   #Pickling
    pickle.dump(feat, fp)
```

```python
X=X.loc[:,feat]
```

```python
X.to_csv('data/X_t0.csv', index=False, header=True)
y.to_csv('data/y_t0.csv', index=False, header=True)
ids.to_csv('data/ids_t0.csv', index=False, header=True)
```
